```python
# adicionando ao dataframe

df = pd.read_csv(r'C:\Users\Rodrigo Correa\projects\Treinamento-Alura-Pandas\aluguel.csv',sep=';',encoding = 'utf-8',
                index_col = False)
```

```python
# definindo lista dos tipos de imoveis para serem filtrados 

lista_tipos_considerar = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão', 'Galpão/Depósito/Armazém', 'Casa Comercial'
                          , 'Terreno Padrão', 'Box/Garagem', 'Loft', 'Loja Shopping/ Ct Comercial'
                          ,'Loteamento/Condomínio', 'Sítio', 'Studio', 'Hotel', 'Indústria']
```

```python
# filtrando conforme lista de imoveis definida

filtro = df['Tipo'].isin(lista_tipos_considerar)
```

```python
# inserindo valores filtrados em um novo df 

df_comercial = df[filtro]
```

```python
# verificando df filtrado

df_comercial
```

```python
# dando nome para coluna do index vazia 

df_comercial.columns.name = 'id'
```

```python
# redefinindo index de 0 ate o tamanho do df 

df_comercial.index = range(0,df_comercial.shape[0])
```

```python
# verificando df 

df_comercial
```

```python
# descubrindo qual valor max, min e media da coluna valor 

valor_max = df_comercial['Valor'].max()
valor_min = df_comercial['Valor'].min()
valor_medio = df_comercial['Valor'].mean()
```

```python
# maiores detelhes do tipo que contem o valor max 

df_comercial.loc[df_comercial['Valor'] == valor_max]
```

```python
# gerando indicadores 

total_de_imoveis_comerciais = df_comercial.shape[0]
total_de_imoveis_acima_da_media = df_comercial.loc[df_comercial['Valor'] > valor_medio,['Tipo']].count()
imoveis_sem_condo = df_comercial['Condominio'].isnull().sum()
```

```python
# bairros com valor mais caro top 10 

df_bairros = df_comercial[['Bairro','Valor','Tipo']]
df_agrupado = df_bairros.groupby(by=['Bairro','Tipo'],as_index =False)['Valor'].sum().sort_values(by='Valor', ascending=False).head(11)
df_agrupado.index = range(0,11)
df_agrupado
```

```python
# imovel com maior area 

max_area = df_comercial['Area'].max()
df_comercial.loc[df_comercial['Area'] == max_area]
```

```python
# top 10 condominios mais caros 

df_condominios = df_comercial[['Tipo','Condominio','Bairro','Area']]
df_condominios = df_condominios.groupby(by=['Tipo','Bairro'],as_index = False).sum().sort_values('Condominio',ascending=False).head(11)
df_condominios.index=range(0,df_condominios.shape[0])
df_condominios
```

```python
df_comercial.sort_values(by=['Condominio'],ascending=False ).head(11)
```

```python
# contando tipos de imoveis

df_tipos_bairro = df_comercial[['Tipo','Bairro']]
df_tipos_bairro = df_comercial.groupby(by=['Bairro'], as_index = False)['Tipo'].value_counts().sort_values(by='count',ascending = False)
df_tipos_bairro.head(16)
```
